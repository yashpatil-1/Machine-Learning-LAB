```python

import pandas as pd
import io
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Function to find the row with largest/smallest value in a column
def find_extremes(df, column):
    max_row = df.loc[df[column].idxmax()]
    min_row = df.loc[df[column].idxmin()]
    return max_row, min_row

# Example usage
max_salary_row, min_salary_row = find_extremes(df, 'salary')
print("Row with max salary:", max_salary_row)
print("Row with min salary:", min_salary_row)

# v. Insert new row with median/mode of each column
def add_summary_row(df):
    summary_row = pd.DataFrame({col: [df[col].median() if pd.api.types.is_numeric_dtype(df[col]) else df[col].mode()[0]] for col in df.columns})
    df = pd.concat([df, summary_row], ignore_index=True)
    return df
df = add_summary_row(df)

# vi. Department-wise sum of salary
dept_salary_sum = df.groupby('department')['salary'].sum()
print("\nDepartment-wise sum of salary:\n", dept_salary_sum)

# vii. Gender-wise largest and lowest salary
gender_salary_extremes = df.groupby('gender')['salary'].agg(['max', 'min'])
print("\nGender-wise salary extremes:\n", gender_salary_extremes)


# viii & ix. Bonus calculation
def calculate_bonus(row):
  if row['salary'] > 50000:
      return 5000
  else:
      return 7000

# Apply bonus based on salary
df['bonus'] = df.apply(calculate_bonus, axis=1)
df['total_compensation'] = df['salary'] + df['bonus']

# x. Plotting graphs (example)

# Histogram of salaries
plt.figure(figsize=(8, 6))
sns.histplot(df['salary'], kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

#Insights: Observe if the salary distribution is skewed or normal. Identify outliers or unusual salary values

# Box plot of salary by department
plt.figure(figsize=(10, 6))
sns.boxplot(x='department', y='salary', data=df)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=45, ha='right')
plt.show()

#Insights: Compare salary ranges between different departments. Detect potential disparities or outliers in specific departments.

#Scatter plot of salary vs experience
plt.figure(figsize=(8,6))
sns.scatterplot(x="Years of Experience", y="salary", data=df)
plt.title('Salary vs Years of Experience')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
#Insights: Observe the relationship between years of experience and salary. A positive correlation indicates that employees with more experience tend to have higher salaries.

# Bar plot for gender wise salary
plt.figure(figsize=(8,6))
sns.barplot(x = 'gender', y = 'salary', data = df)
plt.show()
#Insights: Compares the average salaries between genders.  Note any significant differences.


# Add more plots as per your data

print(df.head()) # Print first few rows to verify changes
```
